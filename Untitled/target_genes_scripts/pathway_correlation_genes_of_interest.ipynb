```python

keywords = {
    'inhibition_tcell': ['inhibition', 'negative regulation', 'T cell activation'],
    'cd8_cytotoxicity': ['CD8', 'cytotoxicity', 'activation'],
    'antigen_presentation': ['antigen presentation', 'MHC'],
    'mhc_ii': ['MHC II', 'presentation'],
    'inflammatory_signaling': ['inflammatory response', 'cytokine', 'inflammation'],
    'tgf_beta_signaling': ['TGF-beta', 'TGF-beta signaling'],
    't_cell_activation_exhaustion': ['T cell activation', 'T cell exhaustion']
}

def filter_pathways(df, keywords_dict):
    filtered_pathways = {}
    for category, terms in keywords_dict.items():
        condition = df['pathway'].str.contains('|'.join(terms), case=False, na=False)
        filtered_pathways[category] = df[condition]
    return filtered_pathways

# Assuming 'sample_data' is your DataFrame and 'pathway' is the column of interest
filtered_pathways = filter_pathways(pathway_activation, keywords)

# Display filtered pathways for each category
for category, pathways in filtered_pathways.items():
    print(f"\nPathways related to {category}:")
    print(pathways['pathway'].unique())
    print(f"Number of pathways in {category}: {len(pathways)}\n")
```

```python
for pathway in pathway_activation_filtered_with_clinical['pathway'].unique():
    stroma_pathway_activation_filtered_with_clinical = pathway_activation_filtered_with_clinical[pathway_activation_filtered_with_clinical['Segment']=='stroma']
    subset= stroma_pathway_activation_filtered_with_clinical[stroma_pathway_activation_filtered_with_clinical['pathway'] == pathway]

    group_low = subset[subset['OS_HL_group'] == 'Low']['ssgsea_score']
    group_high = subset[subset['OS_HL_group'] == 'High']['ssgsea_score']

    stat, p = mannwhitneyu(group_low, group_high, alternative='two-sided')

    print(f"Mann-Whitney U test for {pathway} (low vs high OS): p = {p:.8f}")
   
    subset.head(5)
    plt.figure(figsize=(7, 4))
    sns.boxplot(x="OS_HL_group", y="ssgsea_score", data=subset, color='blue')
    sns.stripplot(x="OS_HL_group", y="ssgsea_score", data=subset, color='grey', alpha=0.2, jitter=True)

    plt.text(
        1.6, # x-position (just to the right of the rightmost box)
        subset['ssgsea_score'].median(),  # y-position (around middle of the plot)
        f"p = {p:.6f}",
        ha='left',
        va='center',
        fontsize=8,
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="black")
    )

    plt.title(f"{pathway} pathway activation in all cells by OS")
    plt.xlabel("OS Group")
    plt.ylabel("Pathway Activation Level")
    plt.show()
```

```python
significant_pathways = []

# Loop through unique pathways
for pathway in pathway_activation_filtered_with_clinical['pathway'].unique():
    stroma_pathway_activation_filtered_with_clinical = pathway_activation_filtered_with_clinical[pathway_activation_filtered_with_clinical['Segment']=='stroma']
    subset = stroma_pathway_activation_filtered_with_clinical[
        stroma_pathway_activation_filtered_with_clinical['pathway'] == pathway
    ]

    group_low = subset[subset['OS_HL_group'] == 'Low']['ssgsea_score']
    group_high = subset[subset['OS_HL_group'] == 'High']['ssgsea_score']

    stat, p = mannwhitneyu(group_low, group_high, alternative='two-sided')

    print(f"Mann-Whitney U test for {pathway} (low vs high OS): p = {p:.4f}")
    if p < 0.05:
        print(f"Significant difference in pathway: {pathway} (p = {p:.9f})")
        significant_pathways.append((pathway, p))
    else:
        print(f"No significant difference in pathway: {pathway} (p = {p:.9f})")
```

To test if the pathway scores follow normal distribution

```python
for pathway in pathway_activation_filtered_with_clinical['pathway'].unique():
    stroma_pathway_activation_filtered_with_clinical = pathway_activation_filtered_with_clinical[pathway_activation_filtered_with_clinical['Segment']=='stroma']
    subset = stroma_pathway_activation_filtered_with_clinical[
        stroma_pathway_activation_filtered_with_clinical['pathway'] == pathway]

    # Group the ssGSEA scores by OS group
    grouped = subset.groupby("OS_HL_group")['ssgsea_score']

    # Run Shapiro-Wilk normality test for each group
    for name, scores in grouped:
        stat, p = shapiro(scores)
        print(f"{pathway} | Group: {name} | p-value = {p:.4f}")

    plt.figure(figsize=(8, 5))
    sns.histplot(data=subset, x="ssgsea_score", hue="OS_HL_group", kde=True, element="step", stat="density",bins=50)
    plt.title(f"{pathway} - Distribution of ssGSEA scores by OS group")
    plt.xlabel("ssGSEA Score")
    plt.ylabel("Density")
    plt.show()
```
