```python
cox_models={}
for gene in target_CAFs:
    cph = CoxPHFitter()
    cph.fit(
        average_fibr_gene_expression_tsi_stroma[[gene, "PFS_months", "PFS_event"]],
        duration_col='PFS_months',
        event_col='PFS_event'
    )
    cox_models[gene] = cph  # store the fitted model by gene name
    print(f"Summary for {gene}")
    #cph.print_summary()
```

```python
cox_models={}
for gene in target_receptors_molecules:
    cph = CoxPHFitter()
    cph.fit(
        average_fibr_gene_expression_tsi_stroma[[gene, "PFS_months", "PFS_event"]],
        duration_col='PFS_months',
        event_col='PFS_event'
    )
    cox_models[gene] = cph  # store the fitted model by gene name
    print(f"Summary for {gene}")
    #cph.print_summary()
```

```python
for gene in target_CAFs:
    cph = cox_models[gene]
    cph.plot(hazard_ratios=True)
    plt.title("Hazard Ratios (Cox Model)")
    plt.tight_layout()
    plt.show()
```

```python
summary_list = []

for gene in target_receptors_molecules:
    cph = cox_models[gene]
    row = cph.summary.loc[gene].copy()
    row["gene"] = gene
    summary_list.append(row)

# Combine into one DataFrame
summary_df = pd.DataFrame(summary_list)
```

```python
summary_df = summary_df.sort_values("exp(coef)")

# Plot
plt.figure(figsize=(8, len(summary_df) * 0.5))
plt.errorbar(
    summary_df["exp(coef)"],           # HR
    summary_df["gene"],                # Gene names (y-axis)
    xerr=[
        summary_df["exp(coef)"] - summary_df["exp(coef) lower 95%"],
        summary_df["exp(coef) upper 95%"] - summary_df["exp(coef)"]
    ],
    fmt='o'
)

plt.axvline(x=1, color='grey', linestyle='--')  # HR = 1 reference line
plt.xlabel("Hazard Ratio (95% CI)")
plt.title("Cox Model Hazard Ratios for CAF Genes")
plt.tight_layout()
plt.show()
```

```python
# After collecting the summaries
summary_df = pd.DataFrame(summary_list)

# Drop CCR5
summary_df = summary_df[summary_df["gene"] != "CCR5"]
summary_df = summary_df[summary_df["gene"] != "CXCR2"]
summary_df = summary_df[summary_df["gene"] != "HYAL2"]

# (Optional) sort for prettier plot
summary_df = summary_df.sort_values("exp(coef)")

# Plot again
import matplotlib.pyplot as plt

plt.figure(figsize=(8, len(summary_df) * 0.5))
plt.errorbar(
    summary_df["exp(coef)"],
    summary_df["gene"],
    xerr=[
        summary_df["exp(coef)"] - summary_df["exp(coef) lower 95%"],
        summary_df["exp(coef) upper 95%"] - summary_df["exp(coef)"]
    ],
    fmt='o'
)

plt.axvline(x=1, color='grey', linestyle='--')
plt.xlabel("Hazard Ratio (95% CI)")
plt.title("Cox Model Hazard Ratios (CAF Genes, CCR5 Removed)")
plt.tight_layout()
plt.show()
```
