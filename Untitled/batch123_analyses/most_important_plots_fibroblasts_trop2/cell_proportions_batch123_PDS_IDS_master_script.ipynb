```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests
from matplotlib.patches import Patch

# ==========================================
# 0. SETUP: Editable SVGs & Output Directory
# ==========================================
# Keep text editable in vector programs
mpl.rcParams['pdf.fonttype'] = 42      
mpl.rcParams['svg.fonttype'] = 'none'

# Define output directory
plot_out_dir = "/Users/nikamikhailava/Desktop/Fibroblasts_signature_derivation/dge_plots"
os.makedirs(plot_out_dir, exist_ok=True)

print("Directory ready. Loading and processing data...")

# ==========================================
# 1. LOAD AND PREP DATA
# ==========================================
cell_proportions = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged_new/bp_res_mid_lvl_ct_updated_ct_fraction_batch_3.csv")
fibr_gene_expression_modified = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged_new/fibroblasts_gene_expression_data_new_batch_three_merged_transposed.csv")
print(f"Initial length: {len(fibr_gene_expression_modified)}")

fibr_gene_expression_modified = pd.merge(
    fibr_gene_expression_modified,
    cell_proportions,
    on="dcc_filename",
    how="inner" 
)

clinical_data = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged/9_eyemt_patient_clinical_data_new.csv")
fibr_gene_expression_modified["Patient"] = fibr_gene_expression_modified["Sample"].str.extract(r'(S\d+)')
fibr_gene_expression_modified = fibr_gene_expression_modified.merge(clinical_data[["Patient","PFS_quartile_b123","OS_quartile_b123"]], on="Patient", how="left")
fibr_gene_expression_cell_abundance = fibr_gene_expression_modified.merge(cell_proportions, on="dcc_filename", how="left")

cell_types = [
    'Tcells_reg', 'Bcells', 'Fibroblasts', 'NKcells', 'Tcells_CD8',
    'Tcells_CD4', 'Macrophages_Monocytes', 'DCs', 'Endothelial_cells'
]

fibr_gene_expression_cell_abundance["PFS_HL_group"] = fibr_gene_expression_cell_abundance["PFS_quartile_b123"].apply(
    lambda x: "Low" if x in [1, "1", 2, "2"] else ("High" if x in [3, "3", 4, "4"] else None)
)
fibr_gene_expression_cell_abundance["OS_HL_group"] = fibr_gene_expression_cell_abundance["OS_quartile_b123"].apply(
    lambda x: "Low" if x in [1, "1", 2, "2"] else ("High" if x in [3, "3", 4, "4"] else None)
)

stroma_cell_abundance = cell_proportions[
    (cell_proportions['Segment'] == 'stroma') & 
    (cell_proportions['Segment_geomx'] == 'stroma')
]

stroma_tsi_cell_abundance = cell_proportions[
    (cell_proportions['Segment'] == 'stroma') & 
    (cell_proportions['Segment_geomx'] == 'tsi')
]

average_stroma_cell_abundance_stroma = stroma_cell_abundance.groupby('Sample')[cell_types].mean()
average_stroma_cell_abundance_stroma["Segment_geomx"] = "stroma"
average_stroma_cell_abundance_tsi = stroma_tsi_cell_abundance.groupby('Sample')[cell_types].mean()
average_stroma_cell_abundance_tsi["Segment_geomx"] = "tsi"

combined = pd.concat([average_stroma_cell_abundance_stroma, average_stroma_cell_abundance_tsi])

# ==========================================
# PLOT VARIATION 1: Sorted by Total Mean (FDR Adjusted)
# ==========================================
print("Generating Plot 1...")
results = {}
for cell in cell_types:
    group_stroma = combined.loc[combined["Segment_geomx"] == "stroma", cell].dropna()
    group_tsi    = combined.loc[combined["Segment_geomx"] == "tsi", cell].dropna()
    stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative="two-sided")
    results[cell] = {
        "stroma_mean": group_stroma.mean(),
        "tsi_mean": group_tsi.mean(),
        "p_value": p_value,
        "mean_total": pd.concat([group_stroma, group_tsi]).mean()
    }

res_df = pd.DataFrame(results).T
res_df["p_adj"] = multipletests(res_df["p_value"], method="fdr_bh")[1]
sorted_cells = res_df.sort_values(by="mean_total", ascending=False).index.tolist()

plot_data = pd.melt(
    combined.loc[combined["Segment_geomx"].isin(["stroma", "tsi"])],
    id_vars=["Segment_geomx"], value_vars=sorted_cells, var_name="Cell_Type", value_name="Expression"
)

plt.figure(figsize=(14, 8))
palette = {"stroma":"#9e82f2", "tsi": "#3ab7be"}
ax = sns.boxplot(x="Cell_Type", y="Expression", hue="Segment_geomx", data=plot_data, palette=palette, linewidth=1.5, order=sorted_cells)

y_max = plot_data["Expression"].max()
for i, cell in enumerate(sorted_cells):
    p_adj = res_df.loc[cell, "p_adj"]
    if p_adj < 0.05:
        star = "*" if p_adj >= 0.01 else "**" if p_adj >= 0.001 else "***"
        ax.text(i, y_max*0.95, star, ha="center", va="center", fontsize=14, fontweight="bold", color="black")

plt.title("Cell Abundance in Distant Stroma vs TSI (FDR-adjusted) in PRE and POST samples", fontsize=16, fontweight="bold")
plt.xlabel("Cell Type")
plt.ylabel("Proportion of Cells")
plt.xticks(rotation=60, fontsize=12)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[:2], labels=["stroma", "tsi"], title="Segments", title_fontsize=13, fontsize=12, loc='upper left', bbox_to_anchor=(1.0, 0.85), frameon=False)
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Cell_Abundance_V1_TotalMeanSorted_FDR.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()

# ==========================================
# PLOT VARIATION 2: Sorted by Stroma Mean (Raw P-Value)
# ==========================================
print("Generating Plot 2...")
sorted_cells_v2 = res_df.sort_values(by="stroma_mean", ascending=False).index.tolist()

plot_data_v2 = pd.melt(
    combined.loc[combined["Segment_geomx"].isin(["stroma", "tsi"])], 
    id_vars=["Segment_geomx"], value_vars=sorted_cells_v2, var_name="Cell_Type", value_name="Expression"
)

plt.figure(figsize=(12, 8))
palette_v2 = {"tsi": "#3ab7be", "stroma": "#9e82f2"}
ax2 = sns.boxplot(x="Cell_Type", y="Expression", hue="Segment_geomx", data=plot_data_v2, palette=palette_v2, linewidth=1.5)

for i, cell in enumerate(sorted_cells_v2):
    p = res_df.loc[cell, "p_value"]
    if p < 0.05:
        star = "*" if p >= 0.01 else "**" if p >= 0.001 else "***"
        ax2.text(i, ax2.get_ylim()[1]*0.95, star, color="black", ha="center", va="center", fontsize=14, fontweight="bold")

plt.title("Cell Abundance in Stromal Segments in Distant Stroma vs TSI", fontsize=16)
plt.xlabel("Cell Type")
plt.xticks(fontsize=12, rotation=60)
plt.ylabel("Proportion of Cells")

custom_legend = [Patch(facecolor="none", edgecolor="none", label=lbl) for lbl in ["* p < 0.05", "** p < 0.01", "*** p < 0.001"]]
ax2.legend(handles=custom_legend, loc='upper left', bbox_to_anchor=(1.05, 1), frameon=False)

handles, labels = ax2.get_legend_handles_labels()
ax2.add_artist(plt.legend(handles=[handles[0], handles[1]], labels=["stroma", "tsi"], loc='upper left', bbox_to_anchor=(1.0, 0.85), title="Segments", title_fontsize=13, fontsize=12))
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Cell_Abundance_V2_StromaMeanSorted_RawP.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()

# ==========================================
# PLOT VARIATION 3: Unsorted (Original List) (FDR Adjusted)
# ==========================================
print("Generating Plot 3...")
plot_data_v3 = pd.melt(
    combined.loc[combined["Segment_geomx"].isin(["stroma", "tsi"])],
    id_vars=["Segment_geomx"], value_vars=cell_types, var_name="Cell_Type", value_name="Expression"
)

plt.figure(figsize=(12, 8))
ax3 = sns.boxplot(x="Cell_Type", y="Expression", hue="Segment_geomx", data=plot_data_v3, palette=palette_v2, linewidth=1.5)

y_max_v3 = ax3.get_ylim()[1]
for i, cell in enumerate(cell_types):
    p_adj = res_df.loc[cell, "p_adj"]
    if p_adj < 0.05:
        star = "*" if p_adj >= 0.01 else "**" if p_adj >= 0.001 else "***"
        ax3.text(i, y_max_v3 * 0.95, star, ha="center", va="center", fontsize=14, fontweight="bold", color="black")

plt.title("Cell Abundance in Distant Stroma vs TSI (FDR-adjusted)", fontsize=16)
plt.xlabel("Cell Type")
plt.ylabel("Expression Level")
plt.xticks(rotation=60, fontsize=12)

handles, labels = ax3.get_legend_handles_labels()
ax3.legend(handles=handles[:2], labels=["stroma", "tsi"], title="Segments", title_fontsize=13, fontsize=12, loc="upper left", bbox_to_anchor=(1.0, 0.85), frameon=False)
signif_legend = [Patch(facecolor="none", edgecolor="none", label=lbl) for lbl in ["* FDR < 0.05", "** FDR < 0.01", "*** FDR < 0.001"]]
ax3.add_artist(plt.legend(handles=signif_legend, loc="upper left", bbox_to_anchor=(1.0, 1.05), frameon=False))
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Cell_Abundance_V3_Unsorted_FDR.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()

# ==========================================
# PLOT VARIATION 4: Total Mean Sorted + Extra Legend (FDR Adjusted)
# ==========================================
print("Generating Plot 4...")
plt.figure(figsize=(14, 8))
ax4 = sns.boxplot(x="Cell_Type", y="Expression", hue="Segment_geomx", data=plot_data, palette=palette, linewidth=1.5, order=sorted_cells)

y_max_v4 = plot_data["Expression"].max()
for i, cell in enumerate(sorted_cells):
    p_adj = res_df.loc[cell, "p_adj"]
    if p_adj < 0.05:
        star = "*" if p_adj >= 0.01 else "**" if p_adj >= 0.001 else "***"
        ax4.text(i, y_max_v4*0.95, star, ha="center", va="center", fontsize=14, fontweight="bold", color="black")

plt.title("Cell Abundance in Distant Stroma vs TSI (FDR-adjusted)", fontsize=16, fontweight="bold")
plt.xlabel("Cell Type")
plt.ylabel("Proportion of Cells")
plt.xticks(rotation=60, fontsize=12)

handles, labels = ax4.get_legend_handles_labels()
ax4.legend(handles=handles[:2], labels=["stroma", "tsi"], title="Segments", title_fontsize=13, fontsize=12, loc="upper left", bbox_to_anchor=(1.0, 0.85), frameon=False)
signif_legend = [Patch(facecolor="none", edgecolor="none", label=lbl) for lbl in ["* FDR < 0.05", "** FDR < 0.01", "*** FDR < 0.001"]]
ax4.add_artist(plt.legend(handles=signif_legend, loc="upper left", bbox_to_anchor=(1.0, 1.05), frameon=False))
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Cell_Abundance_V4_TotalMeanSorted_FDR_WithLegends.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()

print("\n🎉 Done! All 4 SVG plot variations have been saved to the dge_plots folder.")
```
