```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap
import matplotlib as mpl
from pathlib import Path
import os
from scipy.stats import mannwhitneyu, ttest_ind
import scipy.cluster.hierarchy as sch
from statsmodels.stats.multitest import multipletests

# ==========================================
# 0. SETUP: Editable SVGs & Output Directory
# ==========================================
# Tell Matplotlib to keep text editable in vector programs (Affinity/Illustrator)
mpl.rcParams['pdf.fonttype'] = 42      
mpl.rcParams['svg.fonttype'] = 'none'

# Define plot output directory
plot_out_dir = "/Users/nikamikhailava/Desktop/Fibroblasts_signature_derivation/dge_plots"
os.makedirs(plot_out_dir, exist_ok=True)

# Define data output directory for CSVs
data_out_dir = Path("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged_new")
data_out_dir.mkdir(parents=True, exist_ok=True)

print("Directories ready. Loading data...")

# ==========================================
# 1. LOAD AND PREP DATA
# ==========================================
fibr_gene_expression_modified = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged_new/fibroblasts_gene_expression_data_new_batch_three_merged_transposed.csv")
print(f"Initial length: {len(fibr_gene_expression_modified)}")

cell_proportions = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged_new/bp_res_mid_lvl_ct_updated_ct_fraction_batch_3.csv")

fibr_gene_expression_modified = pd.merge(
    fibr_gene_expression_modified,
    cell_proportions,
    on="dcc_filename",
    how="inner" 
)

clinical_data = pd.read_csv("/Users/nikamikhailava/Desktop/Lab_stuff/batch_merged/9_eyemt_patient_clinical_data_new.csv")
fibr_gene_expression_modified["Patient"] = fibr_gene_expression_modified["Sample"].str.extract(r'(S\d+)')

fibr_gene_expression_modified.drop_duplicates(subset=["dcc_filename"], inplace=True)

# ==========================================
# 2. PLOT 1: Stacked Bar Chart (Patients)
# ==========================================
print("Generating Plot 1: Stacked Bar Chart...")
comb_counts = fibr_gene_expression_modified.groupby(['Segment_geomx', 'Segment', 'Patient']).size().reset_index(name='count')
comb_counts = comb_counts[comb_counts["Segment"] == "stroma"]

pivot_df = comb_counts.pivot_table(index='Segment_geomx', columns='Patient', values='count', fill_value=0)

cmap = plt.get_cmap('tab20c')  
colors = [cmap(i % cmap.N) for i in range(len(pivot_df.columns))]  

ax = pivot_df.plot(kind='bar', stacked=True, figsize=(10, 7), color=colors, legend=False)

handles = [Patch(color=color, label=patient) for color, patient in zip(colors, pivot_df.columns)]
ax.legend(handles=handles, title="Patient", bbox_to_anchor=(1.0, 1.1), fontsize=7, title_fontsize=6, loc='upper left')

plt.title("Count of stroma AOIs from TSI and from Distant Stroma, colored by sample", fontsize=16, fontweight='bold')
plt.xlabel("Segment_geomx")
plt.ylabel("Count")
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Stroma_AOIs_Stacked_Barplot_Patients.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()


# ==========================================
# 3. PLOT 2: Manual Bar Chart (Samples)
# ==========================================
print("Generating Plot 2: Manual Bar Chart...")
comb_counts = fibr_gene_expression_modified.groupby(['Segment_geomx', 'Segment', 'Sample']).size().reset_index(name='count')
comb_counts = comb_counts[comb_counts["Segment"] == "stroma"]

pivot_df = comb_counts.pivot_table(index='Segment_geomx', columns='Sample', values='count', fill_value=0)

bar_colors = ["#DB615C", "#65D5DF",]
fig, ax = plt.subplots(figsize=(5, 7))

for i, (idx, row) in enumerate(pivot_df.iterrows()):
    bottom = 0
    for val in row:
        # Fallback to default colors if bar_colors list is shorter than rows
        c = bar_colors[i] if i < len(bar_colors) else None
        ax.bar(idx, val, bottom=bottom, color=c)
        bottom += val

plt.title("Count of stroma AOIs from TSI and from distant stroma", fontsize=13, fontweight='bold')
plt.xlabel("Segment_geomx", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Stroma_AOIs_Manual_Barplot_Samples.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()


# ==========================================
# 4. PLOT 3: Seaborn Barplot
# ==========================================
print("Generating Plot 3: Seaborn Barplot...")
comb_counts.sort_values(by='Sample', inplace=True)
sample_counts = comb_counts['Sample'].value_counts()
single_occurrence_samples = sample_counts[sample_counts == 1].index

df_single_occurrence = comb_counts[comb_counts['Sample'].isin(single_occurrence_samples)]

plt.figure(figsize=(8, 5))
sns.barplot(data=comb_counts, x='Segment_geomx', y='count', hue='Sample', palette='tab20')

plt.title("Count of Segment_geomx types per Sample, when segment is stroma")
plt.xlabel("Segment_geomx")
plt.ylabel("Count")
plt.legend().remove()
plt.tight_layout()

plt.savefig(f"{plot_out_dir}/Stroma_AOIs_Seaborn_Barplot.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()


# ==========================================
# 5. CLUSTERMAP FUNCTION BUILDER
# ==========================================
# (To save space and avoid repeating code 3 times for 500, 200, and 100 genes, 
#  I've created a helper function for your standard clustermaps)

def generate_clustermap(n_genes, filename_suffix, use_custom_colors=False):
    print(f"Generating Clustermap for Top {n_genes} Genes...")
    
    expr_data = (
        fibr_gene_expression_modified[fibr_gene_expression_modified["Segment"] == "stroma"]
        .select_dtypes(include=[float, int])
        .drop(columns=['PFS_quartile_b123', 'OS_quartile_b123'], errors='ignore')
    )

    gene_variance = expr_data.var(axis=0)
    top_genes_list = gene_variance.sort_values(ascending=False).head(n_genes).index
    expr_top = expr_data[top_genes_list]
    
    pd.Series(top_genes_list, name="gene").to_csv(data_out_dir / f"top_variable_{n_genes}_genes_stroma_tsi_pre_post.csv", index=False)

    annot_vars = ["Sample", "Segment_geomx", "NACT_status"] 
    annot_vars = [v for v in annot_vars if v in fibr_gene_expression_modified.columns]
    ann = fibr_gene_expression_modified.loc[expr_top.index, annot_vars].copy()

    if "Segment_geomx" in ann.columns:
        ann["Segment_geomx"] = ann["Segment_geomx"].str.strip()
    if "NACT_status" in ann.columns:
        ann["NACT_status"] = ann["NACT_status"].str.strip()

    color_maps = {}
    for var in annot_vars:
        levels = pd.Categorical(ann[var]).categories
        
        if use_custom_colors and var == "NACT_status":
            custom_palette = {"pre": "orange", "post": "green"}
            color_maps[var] = {lvl: custom_palette.get(lvl, "lightgray") for lvl in levels}
        else:
            palette = sns.color_palette("hls", len(levels))
            color_maps[var] = dict(zip(levels, palette))

    col_colors = pd.DataFrame({var: ann[var].map(color_maps[var]) for var in annot_vars}, index=expr_top.index)

    meta_cols = [c for c in ["dcc_filename", "Segment", "Segment_geomx"] if c in fibr_gene_expression_modified.columns]
    filtered_df = fibr_gene_expression_modified.loc[fibr_gene_expression_modified["Segment"] == "stroma", meta_cols + list(top_genes_list)]
    filtered_df.to_csv(data_out_dir / f"stroma_top{n_genes}_with_metadata_pre_post.csv", index=False)

    g = sns.clustermap(
        expr_top.T, method='complete', metric='correlation', z_score=0,
        col_colors=col_colors, xticklabels=False, yticklabels=False, figsize=(12, 12)
    )

    handles, labels = [], []
    for var in annot_vars:
        for level, color in color_maps[var].items():
            handles.append(plt.Line2D([0], [0], marker='s', linestyle='', markerfacecolor=color, markeredgecolor='none', markersize=8))
            labels.append(f"{var}: {level}")

    g.ax_col_dendrogram.legend(handles, labels, title="Annotations", loc="center left", bbox_to_anchor=(1.15, -1), frameon=False)

    plt.title(f"Clustermap of Top {n_genes} Variable Genes in Stroma Segments (PRE+POST NACT)", loc="center", fontsize=18)
    
    plt.savefig(f"{plot_out_dir}/Clustermap_Top{n_genes}_{filename_suffix}.svg", dpi=300, bbox_inches='tight', format='svg')
    plt.show()
    
    return top_genes_list, filtered_df

# Generate the 500, 200, and 100 Clustermaps
top_genes_500, _ = generate_clustermap(500, "Basic", use_custom_colors=False)
top_genes_200, _ = generate_clustermap(200, "Basic", use_custom_colors=False)
top_genes_100, filtered_df_100 = generate_clustermap(100, "Basic", use_custom_colors=True)


# ==========================================
# 6. DIFFERENTIAL EXPRESSION TESTING
# ==========================================
print("Running Statistical Tests on Top 100 Genes...")
group_col = "Segment_geomx"   
stroma_label = "stroma"
tsi_label = "tsi"

gene_cols = list(top_genes_100)
expr = filtered_df_100[gene_cols]
groups = filtered_df_100[group_col]

results = []
for gene in gene_cols:
    stroma_vals = expr.loc[groups == stroma_label, gene].dropna()
    tsi_vals    = expr.loc[groups == tsi_label, gene].dropna()

    if len(stroma_vals) < 2 or len(tsi_vals) < 2:
        p = np.nan
    else:
        p = ttest_ind(stroma_vals, tsi_vals, equal_var=False).pvalue

    results.append([gene, p, stroma_vals.mean(), tsi_vals.mean()])

res = pd.DataFrame(results, columns=["gene", "p_value", "stroma_mean", "tsi_mean"])
res["padj"] = multipletests(res["p_value"].fillna(1), method="fdr_bh")[1] # Fill NA with 1 for correction
res["significant"] = res["padj"] < 0.05
res["TSI_higher"] = res["tsi_mean"] > res["stroma_mean"]
res["TSI_upregulated"] = res["TSI_higher"] & res["significant"]
res = res.sort_values("p_value")
res.to_csv(data_out_dir / "stroma_vs_tsi_DE_table_top100.csv", index=False)


# ==========================================
# 7. CUSTOM CLUSTERMAP (Top 100)
# ==========================================
print("Generating Plot 7: Custom Clustermap...")
custom_cmap = ListedColormap([
    "#00366a", "#00629b", "#1790c9", "#5dc1f1", "#a8e0fb", 
    "#f5f5f5", "#fbb4b3", "#fa7578", "#eb394b", "#be1624", "#830000"
])
segment_colors = {"tsi": "#3ab7be", "stroma": "#9e82f2"}
nact_colors = {"pre": "#5ebeff", "post": "#ffd468"}

expr_data = fibr_gene_expression_modified[fibr_gene_expression_modified["Segment"] == "stroma"].select_dtypes(include=[float, int]).drop(columns=['PFS_quartile_b123', 'OS_quartile_b123'], errors='ignore')
expr_top = expr_data[top_genes_100]

annot_vars = [v for v in ["Patient", "Segment_geomx", "NACT_status"] if v in fibr_gene_expression_modified.columns]
ann = fibr_gene_expression_modified.loc[expr_top.index, annot_vars].copy()

if "Segment_geomx" in ann.columns: ann["Segment_geomx"] = ann["Segment_geomx"].str.strip()
if "NACT_status" in ann.columns: ann["NACT_status"] = ann["NACT_status"].str.strip()

color_maps = {}
if "Segment_geomx" in ann.columns: color_maps["Segment_geomx"] = ann["Segment_geomx"].map(segment_colors)
if "NACT_status" in ann.columns: color_maps["NACT_status"] = ann["NACT_status"].map(nact_colors)

for var in annot_vars:
    if var not in color_maps:
        levels = pd.Categorical(ann[var]).categories
        color_maps[var] = ann[var].map(dict(zip(levels, sns.color_palette("hls", len(levels)))))

col_colors = pd.DataFrame({var: color_maps[var] for var in annot_vars}, index=expr_top.index)

g = sns.clustermap(
    expr_top.T, method='complete', metric='correlation', z_score=0,
    col_colors=col_colors, xticklabels=False, yticklabels=False, figsize=(12, 12),
    vmin=-2, vmax=4, cmap=custom_cmap
)

handles, labels = [], []
for var in annot_vars:
    if var != "Patient": 
        for level, color in (segment_colors if var=="Segment_geomx" else nact_colors).items():
            handles.append(plt.Line2D([0], [0], marker='s', linestyle='', markerfacecolor=color, markersize=8))
            labels.append(f"{var}: {level}")

g.ax_col_dendrogram.legend(handles, labels, title="Annotations", loc="center left", bbox_to_anchor=(1.02, 0.25), frameon=True, fontsize=11, title_fontsize=13)

plt.title("Clustermap of Top 100 Variable Genes in Stroma AOIs (POST and PRE NACT)", loc="center", fontsize=21, fontweight="bold", pad=20, x=11, y=1.05)

plt.savefig(f"{plot_out_dir}/Clustermap_Top100_Custom_Palette.svg", dpi=300, bbox_inches='tight', format='svg')
plt.show()


# ==========================================
# 8. SIGNIFICANT GENES BOXPLOTS (FacetGrid)
# ==========================================
print("Generating Plot 8: Significant Genes Boxplots...")
df_plot = fibr_gene_expression_modified[
    (fibr_gene_expression_modified["Segment"] == "stroma") &
    (fibr_gene_expression_modified["Segment_geomx"].isin(["stroma", "tsi"]))
][["Segment_geomx"] + list(top_genes_100)]

df_long = df_plot.melt(id_vars="Segment_geomx", value_vars=list(top_genes_100), var_name="gene", value_name="expression")

p_values = {}
mean_expression = {}
for gene in list(top_genes_100):
    group_stroma = df_long[(df_long["Segment_geomx"] == "stroma") & (df_long["gene"] == gene)]["expression"]
    group_tsi = df_long[(df_long["Segment_geomx"] == "tsi") & (df_long["gene"] == gene)]["expression"]
    
    if len(group_stroma) > 0 and len(group_tsi) > 0:
        stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative="two-sided")
        p_values[gene] = p_value
        mean_expression[gene] = {"stroma_mean": group_stroma.mean(), "tsi_mean": group_tsi.mean()}

significant_genes_tsi_higher = [
    gene for gene, p in p_values.items() 
    if p < 0.05 and mean_expression[gene]["tsi_mean"] > mean_expression[gene]["stroma_mean"]
]

df_filtered = df_long[df_long["gene"].isin(significant_genes_tsi_higher)]

if not df_filtered.empty:
    g = sns.FacetGrid(df_filtered, col="gene", col_wrap=5, sharey=False, height=3)
    g.map_dataframe(sns.boxplot, x="Segment_geomx", y="expression", palette="Set2")
    g.set_titles("{col_name}")
    g.set_xlabels("")
    g.set_ylabels("Expression")

    plt.tight_layout()
    plt.savefig(f"{plot_out_dir}/Significant_Genes_Boxplots.svg", dpi=300, bbox_inches='tight', format='svg')
    plt.show()
else:
    print("No significant genes found to plot for the boxplots.")

print("\n🎉 Done! All CSVs saved to Lab_stuff/batch_merged_new and all editable SVGs saved to dge_plots!")
```
