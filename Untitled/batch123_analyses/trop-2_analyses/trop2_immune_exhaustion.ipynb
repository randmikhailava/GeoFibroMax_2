```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# -----------------------------
# 1. Copy main dataframes
# -----------------------------
df_cd8 = cd8_gene_expression_modified_w_cell_proportions.copy()
df_tumor = tumor_gene_expression_modified_w_cell_proportions.copy()

# -----------------------------
# 2. Keep only CD8 exhaustion genes present
# -----------------------------
genes_to_use_cd8 = [g for g in cd8_exhaustion_markers if g in df_cd8.columns]
print(f"CD8 exhaustion genes used: {len(genes_to_use_cd8)}")

# -----------------------------
# 3. Convert to numeric
# -----------------------------
df_cd8[genes_to_use_cd8] = df_cd8[genes_to_use_cd8].apply(pd.to_numeric, errors='coerce')

# -----------------------------
# 4. Gene-wise Z-score → exhaustion signature
# -----------------------------
zscored_cd8 = df_cd8[genes_to_use_cd8].apply(lambda x: (x - x.mean()) / x.std(ddof=0), axis=0)
df_cd8['CD8_exhaustion_score'] = zscored_cd8.mean(axis=1, skipna=True)

# -----------------------------
# 5. Filter tumor segments only
# -----------------------------
df_cd8_tumor = df_cd8[df_cd8['Segment'] == 'tumor'].copy()
df_tumor_sub = df_tumor[df_tumor['Segment'] == 'tumor'].copy()
df_tumor_sub['TACSTD2'] = pd.to_numeric(df_tumor_sub['TACSTD2'], errors='coerce')

# -----------------------------
# 6. Aggregate per sample
# -----------------------------
cd8_per_sample = df_cd8_tumor.groupby('Sample')['CD8_exhaustion_score'].mean().reset_index()
tumor_per_sample = df_tumor_sub.groupby('Sample')['TACSTD2'].mean().reset_index()

# -----------------------------
# 7. Merge per-sample averages
# -----------------------------
df_sample = pd.merge(cd8_per_sample, tumor_per_sample, on='Sample', how='inner').dropna()

# -----------------------------
# 8. Scatter plot with correlation
# -----------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df_sample,
    x='CD8_exhaustion_score',
    y='TACSTD2',
    s=100,
    color="#9e82f2"
)

# Add regression line
sns.regplot(
    data=df_sample,
    x='CD8_exhaustion_score',
    y='TACSTD2',
    scatter=False,
    color="#DB615C",
    line_kws={"linewidth":2, "linestyle":"--"}
)

# Spearman correlation
rho, pval = spearmanr(df_sample['CD8_exhaustion_score'], df_sample['TACSTD2'])

# Annotate correlation
plt.text(
    x=df_sample['CD8_exhaustion_score'].max()*0.35,
    y=df_sample['TACSTD2'].max()*0.98,
    s=f"Spearman r = {rho:.2f}\n p = {pval:.2e}",
    fontsize=12,
    fontweight="bold"
)

plt.xlabel("Mean CD8 Exhaustion Score per Sample (Tumor)")
plt.ylabel("Mean TROP2 Expression per Sample (Tumor)")
plt.title("Correlation of CD8 Exhaustion Signature with TROP2 (Per Sample, Tumor AOIs)")
plt.tight_layout()
plt.show()
```
