```python
fibr_gene_expression_cell_abundance["PFS_HL_group"] = fibr_gene_expression_cell_abundance["PFS_quartile_b12"].apply(
    lambda x: "Low" if x in [1, "1", 2, "2"] else ("High" if x in [3, "3", 4, "4"] else None)
)
fibr_gene_expression_cell_abundance["OS_HL_group"] = fibr_gene_expression_cell_abundance["OS_quartile_b12"].apply(
    lambda x: "Low" if x in [1, "1", 2, "2"] else ("High" if x in [3, "3", 4, "4"] else None)
)
```

```python
fibr_gene_expression_cell_abundance_tsi_stroma=fibr_gene_expression_cell_abundance[(fibr_gene_expression_cell_abundance["Segment_geomx"]=="tsi") & (fibr_gene_expression_cell_abundance["Segment_y"]=="stroma")]
```

```python
sample_metadata= fibr_gene_expression_cell_abundance_tsi_stroma[["Sample_y", "PFS_HL_group", "OS_HL_group"]].drop_duplicates().set_index("Sample_y")
average_tsi_stroma_gene_expression=fibr_gene_expression_cell_abundance_tsi_stroma.groupby('Sample_y')[target_CAFs+target_receptors_molecules].mean()
average_tsi_stroma_gene_expression = average_tsi_stroma_gene_expression.join(sample_metadata)
average_tsi_stroma_gene_expression.head(5)
```

```python
# Determine the number of plots
num_plots = len(target_CAFs)

# Determine the number of rows and columns for subplots
# Here, we'll use a 2-column layout and calculate the number of rows
ncols = 4
nrows = (num_plots + 1) // ncols  # This will ensure an extra row if needed

# Create the subplots grid
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(17, nrows * 6))

# Flatten axes array if there are multiple rows and columns
axes = axes.flatten()

for i, gene in enumerate(target_CAFs):
    group_low = average_tsi_stroma_gene_expression[average_tsi_stroma_gene_expression["PFS_HL_group"] == "Low"][gene]
    group_high = average_tsi_stroma_gene_expression[average_tsi_stroma_gene_expression["PFS_HL_group"] == "High"][gene]

    # Perform the Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_low, group_high, alternative='two-sided')
    
    # Create the plot
    sns.boxplot(x="PFS_HL_group", y=gene, data=average_tsi_stroma_gene_expression, color='lightblue', order=["Low", "High"], ax=axes[i])
    sns.stripplot(x="PFS_HL_group", y=gene, data=average_tsi_stroma_gene_expression, color='grey', alpha=0.6, jitter=True, order=["Low", "High"], ax=axes[i])
    
    # Set titles and labels for each subplot
    axes[i].set_title(f"{gene} expression in fibroblasts in stroma segment by PFS",fontsize=10)
    axes[i].set_xlabel("PFS Group")
    axes[i].set_ylabel("Gene Expression Level")
    
    # Add p-value to the legend
    axes[i].legend([f'Mann-Whitney p-value: {p_value:.4f}'], loc='upper left', fontsize=8)

# Adjust layout to prevent overlap and make sure labels are visible
plt.tight_layout()
plt.show()
```

```python

# Initialize an empty list to store the p-values for each gene
p_value_matrix = []

# Loop over each gene to calculate the Mann-Whitney U test p-values
for gene in target_CAFs:
    # Extract the "Low" and "High" PFS group data
    group_low = average_tsi_stroma_gene_expression[average_tsi_stroma_gene_expression["PFS_HL_group"] == "Low"][gene]
    group_high = average_tsi_stroma_gene_expression[average_tsi_stroma_gene_expression["PFS_HL_group"] == "High"][gene]
    
    # Perform the Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_low, group_high, alternative='two-sided')
    
    # Append the p-value to the list
    p_value_matrix.append(p_value)

# Convert the p_value_matrix to a numpy array for easier manipulation
p_value_matrix = np.array(p_value_matrix).reshape(1, -1)  # Reshape to 1 row and multiple columns (genes)

# Create a heatmap for the p-values with flipped x and y labels
plt.figure(figsize=(9, 6))
sns.heatmap(p_value_matrix.T, annot=True, fmt=".4f", cmap="YlGnBu", cbar_kws={'label': 'Mann-Whitney p-value'},
            xticklabels=["PFS Group Comparison"], yticklabels=target_CAFs, annot_kws={'size': 10})

# Customize the plot
plt.title("Heatmap of Mann-Whitney U Test P-values for Gene Expression Comparisons (Low vs High PFS)", fontsize=13)
plt.xlabel("Comparison (Low vs High PFS)", fontsize=12)
plt.ylabel("Genes (Target CAFs)", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Initialize p_value_matrix and stars_matrix
p_value_matrix = []
stars_matrix = []

# Loop over each gene to calculate the Mann-Whitney U test p-values
for gene in target_CAFs:
    # Extract the "Low" and "High" OS group data
    group_low = fibr_gene_expression_cell_abundance_tsi_stroma[fibr_gene_expression_cell_abundance_tsi_stroma["OS_HL_group"] == "Low"][gene]
    group_high = fibr_gene_expression_cell_abundance_tsi_stroma[fibr_gene_expression_cell_abundance_tsi_stroma["OS_HL_group"] == "High"][gene]
    
    # Perform the Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_low, group_high, alternative='two-sided')
    
    # Append the p-value to the list
    p_value_matrix.append(p_value)
    
    # Assign stars based on p-value thresholds
    if p_value <= 0.001:
        stars_matrix.append("***")
    elif p_value <= 0.01:
        stars_matrix.append("**")
    elif p_value <= 0.05:
        stars_matrix.append("*")
    else:
        stars_matrix.append("")

# Convert the p_value_matrix to a numpy array for easier manipulation
p_value_matrix = np.array(p_value_matrix).reshape(1, -1)  # Reshape to 1 row and multiple columns (genes)
stars_matrix = np.array(stars_matrix).reshape(1, -1)

# Create a heatmap for the p-values with stars indicating significance
plt.figure(figsize=(9, 6))
sns.heatmap(p_value_matrix.T, annot=stars_matrix.T, fmt="", cmap="YlGnBu", cbar_kws={'label': 'Mann-Whitney p-value'},
            xticklabels=["OS Group Comparison"], yticklabels=target_CAFs, annot_kws={'size': 15, 'weight': 'bold'})

# Customize the plot
plt.title("Heatmap of Mann-Whitney U Test P-values for Gene Expression Comparisons (Low vs High OS)", fontsize=13)
plt.xlabel("Comparison (Low vs High OS)", fontsize=12)
plt.ylabel("Genes (Target CAFs)", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Initialize p_value_matrix and stars_matrix
p_value_matrix = []
stars_matrix = []

# Loop over each gene to calculate the Mann-Whitney U test p-values
for gene in target_receptors_molecules + target_CAFs:
    # Extract the "Low" and "High" PFS group data
    group_low = fibr_gene_expression_cell_abundance_tsi_stroma[fibr_gene_expression_cell_abundance_tsi_stroma["PFS_HL_group"] == "Low"][gene]
    group_high = fibr_gene_expression_cell_abundance_tsi_stroma[fibr_gene_expression_cell_abundance_tsi_stroma["PFS_HL_group"] == "High"][gene]
    
    # Perform the Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_low, group_high, alternative='two-sided')
    
    # Append the p-value to the list
    p_value_matrix.append(p_value)
    
    # Assign stars based on p-value thresholds
    if p_value <= 0.001:
        stars_matrix.append("***")
    elif p_value <= 0.01:
        stars_matrix.append("**")
    elif p_value <= 0.05:
        stars_matrix.append("*")
    else:
        stars_matrix.append("")

# Convert the p_value_matrix to a numpy array for easier manipulation
p_value_matrix = np.array(p_value_matrix).reshape(1, -1)  # Reshape to 1 row and multiple columns (genes)
stars_matrix = np.array(stars_matrix).reshape(1, -1)

# Create a heatmap for the p-values with stars indicating significance
plt.figure(figsize=(9, 6))
sns.heatmap(p_value_matrix.T, annot=stars_matrix.T, fmt="", cmap="YlGnBu", cbar_kws={'label': 'Mann-Whitney p-value'},
            xticklabels=["PFS Group Comparison"], yticklabels=target_receptors_molecules + target_CAFs, annot_kws={'size': 10, 'weight': 'bold'})

# Customize the plot
plt.title("Heatmap of Mann-Whitney U Test P-values for Gene Expression Comparisons (Low vs High PFS)", fontsize=13)
plt.xlabel("Comparison (Low vs High PFS)", fontsize=12)
plt.ylabel("Genes (Target CAFs)", fontsize=11)
plt.tight_layout()
plt.show()
```
