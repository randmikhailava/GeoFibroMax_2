```python
gsea_report_tsi_stroma_c3 = gsea_report_tsi_stroma_c3.rename(columns={'NAME': 'Pathway', 'FDR q-val': 'FDR_qval', 'NES': 'NES'})
```

```python
gsea_report_tsi_stroma_c3 = gsea_report_tsi_stroma_c3.sort_values('FDR_qval')
gsea_report_tsi_stroma_c3["Significant"] = gsea_report_tsi_stroma_c3["FDR_qval"] < 0.25

# Set figure size
plt.figure(figsize=(10, 6))

# Bubble plot
'''
sns.scatterplot(
    data=gsea_report_hallmarks,
    x='NES',                          # NES on x-axis
    y='Pathway',                      # Pathways on y-axis
    size='FDR_qval',                  # Bubble size reflects q-value
    hue='FDR_qval',                   # Bubble color also reflects q-value
    palette='viridis',             # Color gradient (lower q = darker)
    sizes=(20, 200),                  # Bubble size range
    legend='brief'
)
'''


sns.scatterplot(
    data=gsea_report_tsi_stroma_c3,
    x='NES',
    y='Pathway',
    size='FDR_qval',
    hue='Significant',  # Now categorical
    palette={True: 'red', False: 'gray'},
    sizes=(20, 200),
    legend='brief'
)
plt.yticks(size=8)  # Set y-ticks size
plt.title('GSEA Results - Bubble Plot (FDR q-values)')
plt.xlabel('Normalized Enrichment Score (NES)')
plt.ylabel('Gene Sets')
plt.gca().invert_yaxis()  # optional: to have most significant pathways on top
plt.tight_layout()
plt.show()
```

```python

# Extract tags percent
def extract_tags(leading_edge):
    if pd.isna(leading_edge):
        return np.nan
    match = re.search(r"tags=(\d+)%", leading_edge)
    if match:
        return int(match.group(1))
    return np.nan

filtered = gsea_report_tsi_stroma_c3[gsea_report_tsi_stroma_c3["FDR_qval"] < 0.05]
filtered["tags_percent"] = filtered["LEADING EDGE"].apply(extract_tags)
filtered = filtered.sort_values("NES", ascending=False)
filtered["log_1/FDR"] = np.log10(1/(filtered["FDR_qval"] + 1e-10))  # avoid log(0)

plt.figure(figsize=(11,7))

# Scale bubble sizes
min_size = 2.5
max_size = 250
sizes = filtered["tags_percent"].fillna(0)
sizes_scaled = min_size + (sizes / 100) * (max_size - min_size)

# Scatter plot with color mapping to FDR
sc = plt.scatter(
    x=filtered['NES'],
    y=filtered['Pathway'],
    s=sizes_scaled,
    c=filtered['log_1/FDR'],
    cmap='viridis',
    edgecolor='black'
)

plt.xlabel('Normalized Enrichment Score (NES)')
plt.ylabel('Gene Sets')
plt.title('GSEA Results - C2 Dataset')
plt.gca().invert_yaxis()
plt.xticks(rotation=45)

# ------------------- 1️⃣ Colorbar for FDR -------------------
# Create inset axes for colorbar on the right
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

ax = plt.gca()
# axes for colorbar: [x0, y0, width, height] in figure coordinates
cax_color = inset_axes(ax,
                       width="4%",   # width of colorbar
                       height="40%",  # height of colorbar
                       loc='upper right',
                       bbox_to_anchor=(0.13, -0.45, 1, 1),
                       bbox_transform=ax.transAxes)

cbar = plt.colorbar(sc, cax=cax_color)
cbar.set_label(r'$\log_{10} \frac{1}{\mathrm{FDR}}$', rotation=270, labelpad=15)

# ------------------- 2️⃣ Bubble size legend -------------------
legend_sizes = [0, 25, 50, 75, 100]
dummy_handles = [plt.scatter([], [], s=min_size + (size/100)*(max_size - min_size),
                             c='gray', edgecolors='black') for size in legend_sizes]

# axes for bubble legend (placed below colorbar)
cax_bubble = inset_axes(ax,
                        width="8%",
                        height="50%",
                        loc='upper right',
                        bbox_to_anchor=(0.18, 0.0, 1, 1),
                        bbox_transform=ax.transAxes)

bubble_legend = cax_bubble.legend(
    handles=dummy_handles,
    labels=[f"{s}%" for s in legend_sizes],
    scatterpoints=1,
    frameon=True,
    title="Enriched genes (%)",
    loc='center',
    labelspacing=1.3 
)

cax_bubble.axis('off')  # hide axes


plt.tight_layout()
plt.show()
```
