```python
import math

num_plots = len(cell_types)
ncols = 3
nrows = math.ceil(num_plots / ncols)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, nrows * 6))
axes = axes.flatten()

for i, cell in enumerate(cell_types):
    combined = combined.reset_index(drop=True)
    ax = axes[i]

    # Extract data
    group_stroma = combined[combined["Segment_geomx"] == "stroma"][cell]
    group_tsi = combined[combined["Segment_geomx"] == "tsi"][cell]
    group_stroma = group_stroma.dropna()
    group_tsi = group_tsi.dropna()

    # Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative='two-sided')

    # Plot
    sns.boxplot(x="Segment_geomx", y=cell, data=combined, ax=ax, palette=["lightgreen", "lightblue"])
    sns.stripplot(x="Segment_geomx", y=cell, data=combined, color='gray', alpha=0.6, jitter=True, ax=ax)

    # Label
    ax.set_title(f"{cell} proportion in stromal segments(p = {p_value:.4f})", fontsize=10)
    ax.set_xlabel("GeoMX Segment")
    ax.set_ylabel("Cell Proportion")

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Store p-values in a dict
pvals = {}

for cell in cell_types:
    group_stroma = combined.loc[combined["Segment_geomx"] == "stroma", cell].dropna()
    group_tsi    = combined.loc[combined["Segment_geomx"] == "tsi", cell].dropna()
    
    # Mann-Whitney test
    stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative="two-sided")
    pvals[cell] = p_value

# Convert to DataFrame for heatmap
pval_df = pd.DataFrame.from_dict(pvals, orient="index", columns=["p_value"])
pval_df["-log10(p)"] = -np.log10(pval_df["p_value"])   # easier to visualize

# Create heatmap matrix (1 row, many columns)
heatmap_data = pval_df[["-log10(p)"]].T

# Plot
plt.figure(figsize=(len(cell_types)*0.6, 3))
ax = sns.heatmap(
    heatmap_data,
    annot=False,
    cmap="Blues",
    cbar_kws={'label': '-log10(p-value)'}
)

# Add stars for significance
for j, cell in enumerate(heatmap_data.columns):
    if pval_df.loc[cell, "p_value"] < 0.05:
        ax.text(j+0.5, 0.5, "*", color="red", ha="center", va="center", fontsize=16)

ax.set_xticklabels(heatmap_data.columns, rotation=45, ha="right")
ax.set_yticklabels(["Significance"], rotation=0)
plt.title("Mann-Whitney U test significance across cell types")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from matplotlib.lines import Line2D
from matplotlib.text import Text
from matplotlib.patches import Patch

# Store results
results = {}
for cell in cell_types:
    group_stroma = combined.loc[combined["Segment_geomx"] == "stroma", cell].dropna()
    group_tsi    = combined.loc[combined["Segment_geomx"] == "tsi", cell].dropna()

    # Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative="two-sided")

    results[cell] = {
        "stroma_mean": group_stroma.mean(),
        "tsi_mean": group_tsi.mean(),
        "p_value": p_value
    }

# Convert to DataFrame
res_df = pd.DataFrame(results).T

# Heatmap data (two rows: stroma & tsi)
heatmap_data = res_df[["stroma_mean", "tsi_mean"]].T
heatmap_data.index = ["Stroma", "TSI"]

# Plot heatmap
plt.figure(figsize=(len(cell_types)*0.6, 4))
ax = sns.heatmap(
    heatmap_data,
    annot=True, fmt=".2f",
    cmap="viridis",
    cbar_kws={'label': 'Mean cell proportion'},
    linewidths=0.5, linecolor="gray"
)

# Add stars for significant results
for j, cell in enumerate(heatmap_data.columns):
    p = res_df.loc[cell, "p_value"]
    if p < 0.05:
        # place star between the two rows (at y=0.5)
        ax.text(j+0.5, 0.4, "*" if p>=0.01 else "**" if p>=0.001 else "***",
                color="white", ha="center", va="center", fontsize=14, fontweight="bold")
        ax.text(j+0.5, 1.4, "*" if p>=0.01 else "**" if p>=0.001 else "***",
                color="white", ha="center", va="center", fontsize=14, fontweight="bold")

# Labels and title
ax.set_xticklabels(heatmap_data.columns, rotation=45, ha="right")
ax.set_yticklabels(heatmap_data.index, rotation=0)
plt.title("Cell proportions in Stroma vs TSI with significance")


custom_legend = [
    Patch(facecolor="none", edgecolor="none", label="*   p < 0.05"),
    Patch(facecolor="none", edgecolor="none", label="**  p < 0.01"),
    Patch(facecolor="none", edgecolor="none", label="*** p < 0.001")
]

ax.legend(handles=custom_legend, loc='upper left', bbox_to_anchor=(1.1, 1.1),frameon=False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from matplotlib.patches import Patch

# Store results
results = {}
for cell in cell_types:
    group_stroma = combined.loc[combined["Segment_geomx"] == "stroma", cell].dropna()
    group_tsi    = combined.loc[combined["Segment_geomx"] == "tsi", cell].dropna()

    # Mann-Whitney U test
    stat, p_value = mannwhitneyu(group_stroma, group_tsi, alternative="two-sided")

    results[cell] = {
        "stroma_mean": group_stroma.mean(),
        "tsi_mean": group_tsi.mean(),
        "p_value": p_value
    }

# Convert to DataFrame
res_df = pd.DataFrame(results).T

# Create a DataFrame for plotting
plot_data = pd.melt(combined.loc[combined["Segment_geomx"].isin(["stroma", "tsi"])], 
                    id_vars=["Segment_geomx"], 
                    value_vars=cell_types, 
                    var_name="Cell_Type", 
                    value_name="Expression")

# Plot boxplots
plt.figure(figsize=(12, 6))

ax = sns.boxplot(
    x="Cell_Type", y="Expression", hue="Segment_geomx", data=plot_data, 
    palette=["lightblue", "salmon"], linewidth=1.5
)

# Add significance stars and combine with segment labels
for i, cell in enumerate(cell_types):
    p = res_df.loc[cell, "p_value"]
    
    # Determine the star based on p-value
    star = "*" if p >= 0.01 else "**" if p >= 0.001 else "***" if p < 0.001 else ""
    
    if star:  # Only add stars if p < 0.05
        # Place star with segment name above the boxplot (adjust the position slightly)
        ax.text(i, ax.get_ylim()[1] * 0.95, f"{labels[i]} {star}",
                color="black", ha="center", va="center", fontsize=14, fontweight="bold")

# Labels and title
plt.title("Cell Type Abundance in Stromal Segments in Distant Stroma vs TSI")
plt.xlabel("Cell Type")
plt.xticks(fontsize=10, rotation=45)
plt.ylabel("Expression Level")

# Create custom legend for significance
custom_legend = [
    Patch(facecolor="none", edgecolor="none", label="*   p < 0.05"),
    Patch(facecolor="none", edgecolor="none", label="**  p < 0.01"),
    Patch(facecolor="none", edgecolor="none", label="*** p < 0.001")
]

# Add legend for the significance stars
ax.legend(handles=custom_legend, loc='upper left', bbox_to_anchor=(1.05, 1), frameon=False)

# Adjust the legend for stroma and tsi (automatically created by hue)
handles, labels = ax.get_legend_handles_labels()
# Remove the 'hue' legend entry for the 'Cell_Type' (because it is already handled in the plot)
handles = [handles[0], handles[1]]  # Stroma and TSI colors
labels = ["Stroma", "TSI"]

# Add the custom color legend
ax.legend(handles=handles, labels=labels, loc='upper left', bbox_to_anchor=(1.05, 0.75), title="Segments")

plt.tight_layout()
plt.show()
```
